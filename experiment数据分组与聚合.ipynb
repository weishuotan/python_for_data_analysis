##  groupby()分组

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'Animal': ['Falcon', 'Falcon',
                              'Parrot', 'Parrot'],
                   'Max Speed': [380., 370., 24., 26.]})
df
```

```python
df.groupby(['Animal']).mean()
```

```python
df = pd.DataFrame({"Key":['C','B','C','A','B','B','A','C','A'],
                   "Data":[2,4,6,8,10,1,14,16,18]})
group_obj=df.groupby(['Key'])
for i in group_obj:
    print(i)
    print(type(i))
```

```python
df2 = pd.DataFrame({'key1': ['A', 'A', 'B', 'B', 'A'],
                   'key2': ['one', 'two', 'one', 'two', 'one'],
                   'data1': [2, 3, 4, 6, 8],
                   'data2': [3, 5, 6, 3, 7]})
group_obj2 = df2.groupby(by = pd.Series(['a', 'b', 'c', 'a', 'b'])) 
for i in group_obj2:                  
    print(i)
    print(type(i))
```

```python
df2
```

```python
num_df = pd.DataFrame({'a': [1, 2, 3, 4, 5],
                    'b': [6, 7, 8, 9, 10],
                    'c': [11, 12, 13, 14, 15],
                    'd': [5, 4, 3, 2, 1],
                    'e': [10, 9, 8, 7, 6]})
num_df
```

```python
by_column = num_df.groupby({'a':'第一组','b':'第二组','c':'第一组',
           'd':'第三组','e':'第二组'}, axis=1)
for i in by_column:
    print(i)
```

```python
df3 = pd.DataFrame({'a': [1, 2, 3, 4, 5],
                   'b': [6, 7, 8, 9, 10],
                   'c': [5, 4, 3, 2, 1]},
                  index=['Sun', 'Jack', 'Alice', 'Helen', 'Job'])
df3
```

```python
groupby_obj3 = df3.groupby(len)    
groupby_obj3.apply(lambda x: x)
```

```python
groupby_obj3.groups
```

```python
groupby_obj3.get_group(3)
```

##  数据聚合

```python
df4 = pd.DataFrame({'key1': ['A', 'A', 'B', 'B', 'A'],
                   'key2': ['one', 'two', 'one', 'two', 'one'],
                   "data1": [2, 3, 4, 6, 8],
                   "data2": [3, 5, np.nan, 3,7]})
df4
```

```python
df4.groupby(['key1']).apply(lambda x:x)
```

```python
data_frame = pd.DataFrame(np.arange(36).reshape((6, 6)),
                       columns=list('abcdef'))
data_frame['key'] = pd.Series(list('aaabbb'), name='key')
data_frame
```

```python
data_group = data_frame.groupby('key')
```

```python
dict([x for x in data_group])['a']
```

```python
dict([x for x in data_group])['b']
```

```python
data_group.agg('sum')
```

```python
def func(arr):
    return arr.max()-arr.min()
data_group.agg(func)
```

```python
data_group.agg([func, sum])
```

```python
data_group.agg([("最大值",max), ("最小值",min)])
```

```python
data_group.agg({'a': 'sum', 'b': 'mean', 'c': func})
```

##  数据转换

```python
df = pd.DataFrame({'a': [0, 1, 6, 10, 3],
                   'b': [1, 2, 7, 11, 4],
                   'c': [2, 3, 8, 12, 4],
                   'd': [3, 4, 9, 13, 5],
                   'e': [4, 5, 10, 14, 3],
                   'key': ['A', 'A', 'B', 'B', 'B']})
df
```

```python
data_group = df.groupby('key').transform('mean')
```

```python
data_group
```

```python
df = pd.DataFrame({'A': [2, 3, 3, 4, 2],
                   'B': [4, 2, 3, 6, 6],
                   'C': [9, 7, 0, 7, 8],
                   'D': [3, 4, 8, 6, 10]})
key = ['one','one','two',' two',' two']
df.groupby(key).transform('mean')
```
