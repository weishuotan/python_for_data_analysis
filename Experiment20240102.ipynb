```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']

data = pd.read_csv(r'./data/scores.csv')

data_structure=data['data_structure']
linear_algebra=data['linear_algebra']
english=data['english']
python=data['python']

def fenbu(scores):
    excellent=0
    pass_=0
    fail=0
    for i in scores:
        if i>=85: excellent+=1
        elif i>=60&i<85: pass_+=1
        else: fail+=1
    return [excellent,pass_,fail]

data1=fenbu(data_structure)
data2=fenbu(linear_algebra)
data3=fenbu(english)
data4=fenbu(python)

excellent_data=[data1[0],data2[0],data3[0],data4[0]]
pass_data=[data1[1],data2[1],data3[1],data4[1]]
fail_data=[data1[2],data2[2],data3[2],data4[2]]
sum_data=[data1[0]+data2[0]+data3[0]+data4[0],data1[1]+data2[1]+data3[1]+data4[1],data1[2]+data2[2]+data3[2]+data4[2]]

kemu = ['data_structure','linear_algebra','english','python']
fenbu_data = ['优秀','及格','不及格']

fig, axes = plt.subplots(2,2)
ax1=axes[0,0]
ax2=axes[0,1]
ax3=axes[1,0]
ax4=axes[1,1]
ax1.pie(excellent_data,labels=kemu,autopct="%1.1f%%")
ax1.set(title='各门科目的优秀百分比')

ax2.pie(pass_data,labels=kemu,autopct="%1.1f%%")
ax2.set(title='各门科目的及格百分比')

ax3.pie(excellent_data,labels=kemu,autopct="%1.1f%%")
ax3.set(title='各门科目的不及格百分比')

ax4.pie(sum_data,labels=fenbu_data,autopct="%1.1f%%")
ax4.set(title='各成绩层次百分比')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']

data = pd.read_csv(r'./data/sales_data.csv')

#计算每个店铺的平均库存数量
df1=data.groupby(by='store')['stock_qty'].mean() 
print(df1)

# 计算每个店铺的平均库存量以及产品的平均销售价格
he=['stock_qty','price']
df2=data.groupby(by='store')[he].mean()
print(df2)

#计算每个店铺的平均库存量和最大库存量
df3 = data.groupby(by='store')['stock_qty'].agg(['mean','max'])
print(df3)

#对聚合结果进行重命名
df4=data.groupby(by='store')['stock_qty'].agg([('平均值','mean'),('最大值','max')])
print(df4)

#计算每个店铺的平均库存量，最大库存量，产品销售的均价，产品销售的最大值
he=['stock_qty','price']
df5=data.groupby(by='store')[he].agg([('平均值','mean'),('最大值','max')])
print(df5)

#计算每个店铺的平均库存量，产品销售的均价
he=['stock_qty','price']
df6=data.groupby(by='store',as_index=False)[he].agg([('平均值','mean')])
print(df6)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']

data = pd.read_csv(r'./data/sales_data.csv')

df1=data.groupby(by='store')['stock_qty'].mean() 

fig, ax = plt.subplots()

plt.bar(df1.index,df1.values,label='')

ax.set(xlabel='商店名称', ylabel='库存量',
       title='商店名与库存之间的关系')

plt.show()
```
