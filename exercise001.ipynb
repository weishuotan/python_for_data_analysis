```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pylab import mpl
mpl.rcParams['font.sans-serif']=['SimHei']
import csv
import math

accid = pd.read_csv('./data/accid1.csv')

accid
```

```python
col1 = ['accidenttime', 'accidentaddr', 'driver1fault', 'driver1responsibility',
       'sex1', 'platenumber1', 'carcolor1', 
        'jxmc1', 'cclzrq1', 
       'clpp1', 'driver1license', 'brith1']
accid1=accid.loc[:,col1]
```

```python
col2 = ['accidenttime', 'accidentaddr', 'driver1fault', 
       'driver2responsibility', 'sex2',
       'platenumber2', 'carcolor2', 'jxmc2', 'cclzrq2',
       'clpp2', 'driver2license', 
       'brith2']
accid2=accid.loc[:,col2]
```

```python
accid1.head()
```

```python
print(accid1.shape)
print(accid2.shape)
```

```python
#修改列名
col=['accidenttime', 'accidentaddr', 'driverfault', 
       'driverresponsibility', 'sex',
       'platenumber', 'carcolor', 'jxmc', 'cclzrq',
       'clpp', 'driverlicense', 
       'brith']
accid1.columns=accid2.columns=col
#合并数据
new_accid=pd.concat([accid1,accid2],axis=0,ignore_index=True)

new_accid.shape
```

```python
new_accid.head()
```

```python
#删除不负责任的司机
new_accid['driverresponsibility'].value_counts()
```

```python
new_accid.drop(new_accid[new_accid['driverresponsibility']=='不负责任'].index,inplace=True)
new_accid.shape
```

```python
new_accid.reset_index(drop=True,inplace=True)
new_accid.head()
```

```python
#处理交通事故异常值
new_accid['driverfault'].value_counts()
```

```python
new_accid.drop(new_accid[new_accid['driverfault']==' '].index,inplace=True)
new_accid.reset_index(drop=True,inplace=True)
```

```python
new_accid.shape
```

```python
#处理驾驶证发放日期异常值
new_accid['cclzrq'].value_counts().sort_index()
```

```python
new_accid.drop(new_accid[new_accid['cclzrq']=='-1'].index,inplace=True)
new_accid.drop(new_accid[new_accid['cclzrq']=='0001-01-01 00:00:00'].index,inplace=True)
new_accid.reset_index(drop=True,inplace=True)
new_accid.shape
```

```python
#处理 出生日期异常值
new_accid['brith'].value_counts()
```

```python
new_accid['brith']=new_accid['brith'].apply(lambda x:x if x[:2]=='19' else '')
```

```python
new_accid.drop(new_accid[new_accid['brith']==''].index,inplace=True)
new_accid.reset_index(drop=True,inplace=True)
```

```python
new_accid.shape
```

```python
new_accid.to_csv('./data/new_accid.csv',index=False)
```
